```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Fertilizer'] = label_encoder.fit_transform(df['Fertilizer'])

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['District_Name'] = label_encoder.fit_transform(df['District_Name'])

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Soil_color'] = label_encoder.fit_transform(df['Soil_color'])

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Crop'] = label_encoder.fit_transform(df['Crop'])

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Link'] = label_encoder.fit_transform(df['Link'])

df.head()
```

```python
x = df.drop(columns=["Fertilizer"])  # features
y = df["Fertilizer"]  # target variable
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
```

```python
ypred= dt.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred))
```
